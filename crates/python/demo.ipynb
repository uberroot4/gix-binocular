```python
import gix_pyo3
import pandas as pd
import io
import numpy as np
```

```python
GIT_DIR = '/Users/rise/Downloads/24s/2024ss-se-pr-group/24ss-se-pr-inso-01/repo'
NO_MERGES = False
```

```python
traverse_commit_graph_result = gix_pyo3.traverse_commit_graph(
    git_dir=GIT_DIR,
    branches=["master"],
    # threads=4,
    # committish='fb900694931dc9de03d9dd065491290d1b814aa0',
    no_merges=NO_MERGES,
)
# print(
#     f"gix_pyo3: {traverse_result}")

commits_df = pd.read_csv(
    io.StringIO(traverse_commit_graph_result), sep=",", na_values=["nan"]
)

commits_df
```

```python
diffs_result = gix_pyo3.get_diffs(
    git_dir=GIT_DIR,
    threads=4,
    # committish='fb900694931dc9de03d9dd065491290d1b814aa0',
    no_merges=NO_MERGES)
# print(
#     f"gix_pyo3: {traverse_result}")

diffs_df = pd.read_csv(io.StringIO(diffs_result), sep=",", na_values=["nan"])

diffs_df
```

```python
# commits_df["parents"].str.split(';', expand=True)
commits_df["parent"] = commits_df["parents"].apply(lambda g: str(g).split(";"))
df = commits_df.explode("parent")
df["parent"] = df["parent"].str.strip()
df["parent"] = df["parent"].replace("nan", np.NaN)
df
```

```python
df[df['parent'].isna()]
```

```python
df_merged = pd.merge(
    left=df, 
    right=diffs_df,
    how='left',
    left_on=['commit', 'parent'],
    right_on=['commit', 'parent'],
)
df_merged
```

```python
df_merged.to_json('temp.json', orient='records', lines=True)
```
